```python
font_path = '/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf'
font_prop = matplotlib.font_manager.FontProperties(fname=font_path)

df_greeting = df[(df['ym:pv:URL'] == 'http://localhost/greeting') |
                 (df['ym:pv:URL'] == 'capacitor://localhost/greeting') |
                 (df['ym:pv:URL'] == 'https://mts-clatch.morizolabs.ru/greeting')]

df_onboarding = df[df['ym:pv:URL'] == 'capacitor://localhost/onboarding']
df_home = df[(df['ym:pv:URL'] == 'capacitor://localhost/Home') | (df['ym:pv:URL'] == 'http://localhost/home')]

# Получение числа пользователей для каждого экрана
num_users_greeting = len(df_greeting)
num_users_onboarding = len(df_onboarding)
num_users_home = len(df_home)


# Построение гистограммы
screens = ['Экран приветствия', 'Экран анбординга', 'Home']
num_users = [num_users_greeting, num_users_onboarding, num_users_home]
main_colors = ["#56B4E9", "#009E73", '#D55E00']


plt.bar(screens, num_users, color=main_colors)

plt.xticks(fontproperties=font_prop)

# Добавление подписей осей и заголовка
plt.title('Гистограмма пользователей по экранам',fontproperties=font_prop)
plt.xlabel('Воронка регистраций за 25 мая 2023', fontproperties=font_prop)
plt.ylabel('', fontproperties=font_prop)


# Отображение гистограммы
plt.show()
```

```python
import plotly.express as px
data = dict(values=[num_users_greeting, num_users_onboarding, num_users_home],
            labels=['Экран приветствия', 'Экран анбординга', 'Home'])
fig = px.funnel(data, y='labels', x='values')
fig.update_layout(
    title={
        'text': 'Воронка регистраций за 25 мая 2023',
        'x': 0.52,
        'y': 0.92
    }
)
fig.show()
```
